```python
fig, axes = plt.subplots(2, 2, sharey=False, sharex=False)
fig.set_figwidth(14)
fig.set_figheight(8)
axes[0][0].plot(series.index, series['Sales'], label='Original')
axes[0][0].plot(series.index, series['Sales'].rolling(window=2).mean(), label='2-Months Rolling Mean')
axes[0][0].set_xlabel("Years")
axes[0][0].set_ylabel("Sales")
axes[0][0].set_title("3-Months Moving Average")
axes[0][0].legend(loc='best')
axes[0][1].plot(series.index, series['Sales'], label='Original')
axes[0][1].plot(series.index, series['Sales'].rolling(window=3).mean(), label='3-Months Rolling Mean')
axes[0][1].set_xlabel("Years")
axes[0][1].set_ylabel("Sales")
axes[0][1].set_title("3-Months Moving Average")
axes[0][1].legend(loc='best')
axes[1][0].plot(series.index, series['Sales'], label='Original')
axes[1][0].plot(series.index, series['Sales'].rolling(window=8).mean(), label='6-Months Rolling Mean')
axes[1][0].set_xlabel("Years")
axes[1][0].set_ylabel("Sales")
axes[1][0].set_title("6-Months Moving Average")
axes[1][0].legend(loc='best')
axes[1][1].plot(series.index, series['Sales'], label='Original')
axes[1][1].plot(series.index, series['Sales'].rolling(window=8).mean(), label='8-Months Rolling Mean')
axes[1][1].set_xlabel("Years")
axes[1][1].set_ylabel("Sales")
axes[1][1].set_title("8-Months Moving Average")
axes[1][1].legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 1, sharey=False, sharex=False)
fig.set_figwidth(14)
fig.set_figheight(8)


axes.plot(series.index, series['Sales'], label='Original')
axes.plot(series.index, series['Sales'].rolling(window=12).mean(), label='12-Months Rolling Mean')
axes.set_xlabel("Years")
axes.set_ylabel("Sales")
axes.set_title("12-Months Moving Average")
axes.legend(loc='best')
plt.tight_layout()
plt.show()
```
